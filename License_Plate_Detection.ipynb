```
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```
from google.colab.patches import cv2_imshow
```

```
img=cv2.imread("/content/Car-Number-Plate.jpg")
```

```
print(cv2_imshow(img))
```

```
img.shape
```

```
img1=cv2.imread("/content/Car-Number-Plate.jpg",0)
print(cv2_imshow(img1))
```

```
#license plate detection
license_cascade=cv2.CascadeClassifier("/content/haarcascade_russian_plate_number.xml")
```

```
license_plate=license_cascade.detectMultiScale(img1,1.1,4)
```

```
for(x,y,w,h) in license_plate:
  cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)

#displaying our output
cv2_imshow(img)
cv2.waitKey()
```
